```python
for dirname in os.listdir('./ARC/'):
    path1 = './ARC/' + dirname

#     print(path1)
    if os.path.isdir(path1):
        if 'ARC' in dirname:
            num = dirname[-3:]
            path2 = './ARC/' + num

            print(path1)
            print(path2)
            os.rename(path1, path2)
```

```python
heisei = ['a', 'b', 'c', 'd']
reiwa = ['a', 'b', 'c', 'd', 'e', 'f']

for i in range(1, 180):
    if str(i).zfill(3) not in os.listdir('./ABC/'):
        os.mkdir('./ABC/' + str(i).zfill(3))
        if i > 125:
            alphs = reiwa                
        else:
            alphs = heisei
        for alph in alphs:
            print('./ABC/' + str(i).zfill(3) + '/' +  alph + '.py')
            f = open('./ABC/' + str(i).zfill(3) + '/' + alph + '.py', 'w')
            f.close()
```
